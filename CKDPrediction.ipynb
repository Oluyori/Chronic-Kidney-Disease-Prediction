# CKD prediction using Machine Learning Techniques such as DT, KNN, NB, RF, Logreg, SVM, and ANN(MLP)

Biosciences has progressed to a greater degree, and Electronic Health Records have created vast volumes of data. As a result, there is a strong need to generate information from such vast amounts of data. In this area of biosciences, data mining and machine learning are extremely important in predicting and diagnosing diseases such as chronic kidney disease (CKD). CKD is a disorder whereby the kidneys have been damaged and loss its function of filtering blood as efficiently as they once did. This ailment is one of the most serious illnesses worldwide, and it requires prompt diagnosis. Kidney disease or failure in the family, high blood pressure, 2 diabetes mellitus (t2dm) are all risk factors for CKD. This is a long-term kidney problem with a high risk of growing worse over time. Heart disease, anaemia, bone disease, excessive potassium, and calcium levels are all significant consequences of kidney failure. In the worst-case scenario, total renal failure occurs, necessitating a kidney transplant to survive. Early CKD detection can enhance one's quality of life significantly. This necessitates the development of a reliable prediction system to detect CKD at an early stage. A broad range of machine learning techniques have been utilised to predict CKD in the literature to support clinician in early diagnosis. This study reveals the impact of hypertension, diabetes, and other features for prediction of CKD. Multilayer Perceptron, Random Forest, Logistics Regression, Decision Tree, Support Vector Machine, K-Nearest Neighbour and Naïve Bayes are the machine learning techniques applied in this study. Random Forest achieved the highest accuracy of 99% with AUC score of 0.940 upon validation with confusion matrix and Area under ROC curve respectively. Comparison of accuracy among the applied machine learning techniques was evaluated using AUC score and confusion matrix.  This paper has explored the prediction of CKD from this perspective. The CKD dataset was retrieved from the University of California, Irvine (UCI) repository. This study predicts CKD using data preparation, data transformation, and several classifiers, as well as proposing the optimal CKD prediction framework. The framework's findings indicate that a higher prediction of CKD at an early stage is possible. The code for this study is available in my public repository
